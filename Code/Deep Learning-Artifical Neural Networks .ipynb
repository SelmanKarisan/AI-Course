# Regression with deep learning

```python
import pandas as pd 

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df=pd.read_pickle('kc_house.pkl')
```

```python
df.head()
```

```python
df.to_csv('kc.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('price',axis=1)
y=df[['price']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80,random_state=42)
```

```python
model=Sequential()
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1))
model.compile(loss='mse',optimizer='adam')
```

```python
history=model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=150,batch_size=128,verbose=1)
```

```python
model.summary()
```

```python
model.evaluate(x_test,y_test)
```

```python
tahmin=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(tahmin,y_test)
```

```python
#eğitme anındakileri grafik olarak veriyor 
loss_df=pd.DataFrame(history.history)
loss_df.plot(figsize=(12,8))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
(mean_squared_error(y_test,tahmin))**0.5
```
