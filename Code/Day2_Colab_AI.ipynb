```python
import pandas as pd
```

```python
df=pd.read_csv("corona_dat.csv")
```

```python
df.head()
```

```python
df2=df[['date', 'China', 'Italy', 'Brazil', 'Spain', 'United Kingdom', 'US', 'France', 'Turkey']]
```

```python
df2.set_index('date',inplace=True)
```

```python
total=df2.cumsum(axis=0)
```

```python
import bar_chart_race as bcr
```

```python
bcr.bar_chart_race(total,filename="covid.mp4",figsize=(8,5),title="Selman")
```
