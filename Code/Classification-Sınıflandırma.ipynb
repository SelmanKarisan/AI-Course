# Feature Engineering 

```python
# dictionary: key:value
d={"Yes":1,'No':0}
```

```python
df['Attrition']=df['Attrition'].map(d) 
#for d√∂g√ºs√ºz deƒüerleri ≈üak diye deƒüi≈ütirdik
```

```python
# df['Attrition']=df['Attrition'].astype("int64") 
# bunu bo≈ülarƒ± doldurduktan sonra yapƒ±caz
```

```python
bos=df[df['Attrition'].isnull()]
```

```python
dolu=df[df['Attrition'].notnull()]
```

```python
dolu.isnull().sum()
```

```python
x=dolu.drop('Attrition',axis=1)
y=dolu['Attrition']
```

```python
x.head()
```

```python
x['BusinessTravel'].value_counts()
```

```python
x=pd.get_dummies(x,drop_first=True) 
# b√ºt√ºn yazƒ±larƒ± sayƒ± yaptƒ±k 
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x,y)
b.fit(x,y)
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
# classfication Accuracy , Precision , recall f1 score ve confusion matrix 
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report 
```

```python
accuracy_score(tahmin1,y)
```

```python
confusion_matrix(tahmin1,y)
```

```python
print(classification_report(tahmin1,y))
```

```python
sns.heatmap(confusion_matrix(tahmin1,y),annot=True)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.countplot(df['species']) # error verdi hocam √ß√∂zemedim 
```

```python
sns.pairplot(df,hue="species")
```

```python
from sklearn.svm import SVC
```

```python
s=SVC()
```

```python
x=df.drop("species",axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
pred=s.fit(x_train,y_train).predict(x_test)
```

```python
confusion_matrix(y_test,pred) #%100 doƒüru tahmin etti 
```
